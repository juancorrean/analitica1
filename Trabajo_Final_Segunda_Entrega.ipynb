# **5. Transformaciones: Aplicación de funciones para la creación de nuevas variables**

```python
# Se confirma que los datos a sumar para crear una nueva columna que contenga el total de personas involucrados en dichos eventos sean compatibles
datos.dtypes 
```

```python
# Se crea una columna con los resultados de las columnas heridos, fallecidos y desaparecidos, llamada personas involucradas
# y que de como resultado en cada fila (SI ó NO).
# de esta manera se observa si en el evento ocurrido, existieron personas involucradas 

tranformacion = [ 
    ((base_datos.loc[:,'DESAPARECIDOS']  + base_datos.loc[:,'HERIDOS'] + base_datos.loc[:,'FALLECIDOS']) > 0 ),
    ((base_datos.loc[:,'DESAPARECIDOS']  + base_datos.loc[:,'HERIDOS'] + base_datos.loc[:,'FALLECIDOS']) <= 0)]
    
values1 = ['SI','NO']
base_datos['PERSONAS_INVOLUCRADAS']= np.select(tranformacion, values1)
base_datos
```

```python
# Teniendo en cuenta el numero de personas afectadas, se crea una columna que se llama (NIVEL_DEL_INCIDENTE_NATURAL) y 
# que da como resultado de:
# 0 a 100 afectados (LEVE)
# 101 a 1000 afectados (IMPORTANTE)
# 1001 a 10000 afectados( GRAVE) 
# superior (URGENTE)

tranformacion2 = [ 
    ((base_datos.loc[:,'AFECTADOS']  ) <= 100 ),
    (base_datos.loc[:,'AFECTADOS']   > 101) & (base_datos.loc[:,'AFECTADOS']   <= 1000),
    (base_datos.loc[:,'AFECTADOS']   > 1001) & (base_datos.loc[:,'AFECTADOS']  <= 10000),
    (base_datos.loc[:,'AFECTADOS']   > 10001 )
    ]
    
values12 = ['LEVE','IMPORTANTE','GRAVE','URGENTE']
base_datos['NIVEL_DEL_INCIDENTE_NATURAL']= np.select(tranformacion2, values12)
base_datos
```

###**- Pregunta 4: ¿Cuál es el tipo evento que tiene mas afectados y en qué porcentaje?**

```python
# importar paquete
import plotly.graph_objects as go 

# crear base
df0 = base_datos.groupby(['TIPO_DE_EVENTO'])[['HERIDOS']].sum().sort_values('HERIDOS', ascending = False).rename(columns={'HERIDOS':'counts'})
df0['ratio'] = df0.apply(lambda x: x.cumsum()/df0['counts'].sum())

# definir figura
fig = go.Figure([go.Bar(x=df0.index, y=df0['counts'], yaxis='y1', name='sessions id'),
                 go.Scatter(x=df0.index, y=df0['ratio'], yaxis='y2', name='Heridos', hovertemplate='%{y:.1%}', marker={'color': '#000000'})])

# agregar detalles
fig.update_layout(template='plotly_white', showlegend=False, hovermode='x', bargap=.3,
                  title={'text': '<b> Heridos por tipo de evento<b>', 'x': .5}, 
                  yaxis={'title': 'Heridos'},
                  yaxis2={'rangemode': "tozero", 'overlaying': 'y', 'position': 1, 'side': 'right', 'title': 'ratio', 'tickvals': np.arange(0, 1.1, .2), 'tickmode': 'array', 'ticktext': [str(i) + '%' for i in range(0, 101, 20)]})

fig.show()
```

**Interpretación:** Según el gráfico de pareto, los eventos de accidente transporte terrestre, incendio estructural y explosión son los principales responsables del gran número de personas heridas que se analizó en la gráfica anterior. ###**- Pregunta 8: ¿Cuál es la relación de las viviendas averiadas y destruidas? Muestre los afectados según el departamento**

```python
# Se agrupa por evento y departamento, luegos se suman las columnas de viviendas destruidas, averiadas y afectados
df0=base_datos.groupby(['TIPO_DE_EVENTO','DEPARTAMENTO']).agg(
    destruidas = ('VIVIENDAS_DESTRUIDAS','sum'),
    averiadas = ('VIVIENDAS_AVERIADAS','sum'),
    afectados = ('AFECTADOS','sum')
    ).reset_index().sort_values('afectados', ascending = False).head(10)
df0 = df0[df0['TIPO_DE_EVENTO']=='INUNDACION'] # Filtrar DataFrame

# definir gráfica
fig = px.scatter(df0, x = 'destruidas', y ='averiadas', color = 'DEPARTAMENTO', size = 'afectados')

# añadir atributos
fig.update_layout(
    xaxis_title = '<b>Viviendas destruidas<b>',
    yaxis_title = '<b>Viviendas averiadas<b>',
    template = 'simple_white',
    title_x = 0.5,
    legend_title = '<b> Departamento <b>')

fig.show()
```

###**- Pregunta 9: Top 7 de los departamentos con más personas involucradas**
