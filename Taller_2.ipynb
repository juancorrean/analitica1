**Descripción de las variables**
- ARREST_KEY: Identificación de cada arresto.
- ARREST_DATE: Fecha del arresto.
- PD_CD: Código de clasificación interna.
- PD_DESC: Descripción de la clasificación interna correspondiente al código PD.
- KY_CD: Código de clasificación interna de tres dígitos (categoría más general que el código PD).
- OFNS_DESC: Descripción de la clasificación interna correspondiente al código KY (categoría más general que la descripción PD).
- LAW_CODE: Cargos del código de ley correspondientes a la Ley Penal del Estado de Nueva York, VTL y otras diversas leyes locales.
- LAW_CAT_CD: Nivel de ofensa: delito grave, delito menor, violación.
- ARREST_BORO: Municipio de detención. B (Bronx), S (Staten Island), K (Brooklyn), M (Manhattan), Q (Queens).
- ARREST_PRECINCT: Precinto donde ocurrió la detención.
- JURISDICTION_CODE: Jurisdicción responsable de la detención. Los códigos de jurisdicción 0 (Patrulla), 1 (Tránsito) y 2 (Vivienda) representan a la policía de Nueva York, mientras que los códigos 3 y más representan jurisdicciones que no son de la policía de Nueva York.
- GRUPO DE EDAD: Edad del perpetrador dentro de una categoría.
- PERP_SEX: Descripción del sexo del perpetrador.
- PERP_RACE: Descripción de la carrera del perpetrador.
- X_COORD_CD: Coordenada X de bloque medio para el sistema de coordenadas planas del estado de Nueva York, zona de Long Island, NAD 83, unidades pies (FIPS 3104).
- Y_COORD_CD: Coordenada Y de bloque medio para el sistema de coordenadas planas del estado de Nueva York, zona de Long Island, NAD 83, unidades pies (FIPS 3104).
- Latitud: Coordenada de latitud para el Sistema de coordenadas global.
- Longitud: Coordenada de longitud para el Sistema de coordenadas global.
- Lon_Lat: Columna de puntos georreferenciados basada en campos de longitud y latitud.

# **Pregunta 1**


**¿Cuál ha sido la evolución diaria de la cantidad de arrestos?**
- a. a nivel general
- b. a nivel de ofensa (columna del nivel de ofensa: LAW_CAT_CD)
- c. a nivel de barrio (columna del barrio ARREST_BORO)

###*a) a nivel general*

```python
#Se crea un DataFrame con todas las fechas para hacer la serie de tiempo
Seriefecha= pd.DataFrame(pd.date_range(start=Basedatos['FECHA'].min(), end=Basedatos['FECHA'].max())).rename(columns ={0:'FECHA'})
Seriefecha
```

```python
#Se unió la base de crimenes con las fechas
Seriefechacrimenes= pd.merge(Seriefecha, Basedatos, on ='FECHA', how ='left')
```

```python
#Se hizo un groupby para contar los arrestos por día
Serie=Seriefechacrimenes.groupby(['FECHA'])[['ARREST_KEY']].count().reset_index()
Serie.head()
```

```python
#Antes de graficar se suaviza la serie
Serie2=Serie.copy()
Serie2.iloc[:,1:] = Serie2.rolling(window=30).mean()
```

```python
fig = px.line(Serie2, x='FECHA', y ='ARREST_KEY', title = '<b>Evolución arrestos a nivel general<b>',
              color_discrete_sequence=px.colors.qualitative.Safe)

fig.update_layout(
    xaxis_title = 'Año',
    yaxis_title = 'Arrestos',
    template = 'simple_white',
    legend_title = '<b>Nivel de ofensa<b>',
    title_x = 0.5)
fig.show()
```

**Interpretación:** La cantidad general de arrestos tuvo el pico más grande entre el 2006 y el 2014, después de este año los arrestos empezaron a disminuir de manera constante antes de iniciar el año 2020 y finalmente se observa que desde el 2020, el número de arrestos ha ido incrementandose nuevamente.

###*b) a nivel de ofensa (columna del nivel de ofensa: LAW_CAT_CD)*

```python
Basedatos['LAW_CAT_CD'].value_counts()

#Nivel de ofensa:
#felony: delito grave
#misdemeanor: delito menor
#violation: violación
#i: informático
```

```python
#Se hizo un groupby para contar los arrestos a nivel de ofensa
ofensa=Seriefechacrimenes.groupby(['FECHA','LAW_CAT_CD'])[['ARREST_KEY']].count().reset_index()
ofensa.head()
```

```python
#Se filtra por la ofensa Misdemeanor/delito menor
Misdemeanor=ofensa[ofensa['LAW_CAT_CD']=='M']

#Se crea una serie de tiempo nueva con el delito Misdemeanor 
ofensa1= pd.merge(Seriefecha,Misdemeanor, on ='FECHA', how ='left').rename(columns={'ARREST_KEY':'Misdemeanor'})
ofensa1.head()
```

```python
#Se filtra por violación
Violacion=ofensa[ofensa['LAW_CAT_CD']=='V']
Violacion.head()

#Se crea una serie de tiempo para violación
ofensa2= pd.merge(Seriefecha,Violacion, on ='FECHA', how ='left').rename(columns={'ARREST_KEY':'Violacion'})
ofensa2.head()
```

```python
#Se filtra por delito grave
felony=ofensa[ofensa['LAW_CAT_CD']=='F']
felony.head()

#Serie de tiempo para delito grave
ofensa3= pd.merge(Seriefecha,felony, on ='FECHA', how ='left').rename(columns={'ARREST_KEY':'felony'})
ofensa3.head()
```

```python
#se filtra por delito informatico
informatica=ofensa[ofensa['LAW_CAT_CD']=='I']
informatica.head()

#serie de tiempo para delito informatico
ofensa4= pd.merge(Seriefecha,informatica, on ='FECHA', how ='left').rename(columns={'ARREST_KEY':'informaticos'})
ofensa4.head()
```

```python
#Se crea un dataframe nuevo con 5 columnas correspondientes a la cantidad de arrestos por cada tipo de delito
puntob = Seriefecha.copy()
puntob['Delito menor'] = ofensa1['Misdemeanor'].fillna(0)
puntob['Violacion'] = ofensa2['Violacion'].fillna(0)
puntob['Delito mayor'] = ofensa3['felony'].fillna(0)
puntob['Informatica'] = ofensa4['informaticos'].fillna(0)
puntob.head()
```

```python
#Antes de graficar se suaviza la serie
puntob2=puntob.copy()
puntob2.iloc[:,1:] = puntob2.rolling(window=30).mean()
```

```python
fig = px.line(puntob2, x='FECHA', y =['Delito menor','Violacion','Delito mayor','Informatica'], title = '<b>Evolución arrestos por nivel de ofensa<b>',
              color_discrete_sequence=px.colors.qualitative.Bold)

fig.update_layout(
    xaxis_title = 'Año',
    yaxis_title = 'Arrestos',
    template = 'simple_white',
    legend_title = '<b>Nivel de ofensa<b>',
    title_x = 0.5)
fig.show()
```

**Interpretación:** El mayor nivel de ofensa que se presentan en los arrestos es del de delito menor (Misdemeanor) y este ha ido disminuyendo a partir del año 2015 aproximadamente hasta alcanzar a el delito mayor (Felony) en el año 2020, el cual se ha mantenido constante durante el tiempo. El tipo de ofensa violación, junto con informática no han crecido tan significativamente como los demás y son muy pocos los arrestos por estos dos.
